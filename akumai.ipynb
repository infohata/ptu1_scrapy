```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.karkkainen.com/verkkokauppa/rakennus-ja-remontti/sahkotarvike/'
source = requests.get(url).text
soup = BeautifulSoup(source, 'html.parser')
products = soup.body.find_all('div', class_="product")
products
```

```python
prodlist = []
for product in products:
    product_name = product.find('h2', class_='product_name')
    if product_name:
        product_name = product_name.a['title']
    price = product.find('input')
    if price:
        price = price['value'].replace('\xa0', '').split()[0]
    prodlist.append((product_name, price))
prodlist
```

```python
json_list = []
for product in products:
    jsondata = product.find('div', class_='ga4-product-impression-data')
    if hasattr(jsondata, 'data-json'):
        json_list.append(jsondata['data-json'])

json_list
```

```python
import json
product_dicts = []
for json_item in json_list:
    product_dicts.append(json.loads(json_item))

product_dicts
```
