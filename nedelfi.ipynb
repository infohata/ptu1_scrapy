```python
from bs4 import BeautifulSoup
import requests

url = 'https://books.toscrape.com/'
source = requests.get(url).text
source
```

```python
soup = BeautifulSoup(source, 'html.parser')
body = soup.body
body
```

```python
saito_pavadinimas = body.find('div', class_= 'h1').text.strip()
saito_pavadinimas
```

```python
kategorijos = body.find('ul', class_='nav-list')
kategoriju_li = kategorijos.find_all('li')
kategoriju_li[3]
```

```python
nuorodos = {}
for kategorija in kategoriju_li[1:]:
    # nuorodos[kategorija.text.strip()] = kategorija.find('a')['href']
    nuorodos[kategorija.text.strip()] = kategorija.a['href']
nuorodos
```

```python
class Knyga:
    def __init__(self, title, rating, price, in_stock, link, cover_image_link):
        self.title = title
        self.rating = rating
        self.price = price
        self.in_stock = in_stock
        self.link = link
        self.cover_image_link = cover_image_link

    def __repr__(self):
        return f'{self.title}: {self.link}'
```

```python
books_ol = body.find('ol', class_='row')
book_soup = books_ol.find_all('li')
book_soup
```

```python
knygos = []
def add_books(knygos, book_soup):
    for book in book_soup:
        title = book.h3.a['title']
        rating = book.find('p', class_='star-rating')['class'][1]
        price = book.find('p', class_='price_color').text[1:]
        in_stock = (book.find('p', class_='instock').i['class'][0] == 'icon-ok')
        link = book.a['href']
        cover_image_link = book.img['src']
        knyga = Knyga(title, rating, price, in_stock, link, cover_image_link)
        if not knyga in knygos:
            knygos.append(knyga)
    return knygos
knygos = add_books(knygos, book_soup)
knygos
```

```python
pager = body.find('ul', class_='pager')
next = pager.find('li', class_='next')
if next and next.a:
    next_url = url + next.a['href']
next_url
```

```python
next_source = requests.get(next_url).text
next_soup = BeautifulSoup(next_source, 'html.parser')
next_soup.body
```

```python
books_ol = next_soup.body.find('ol', class_='row')
book_soup = books_ol.find_all('li')
knygos2 = add_books(knygos, book_soup)
knygos2
```
